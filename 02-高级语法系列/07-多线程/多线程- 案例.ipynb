```python
# 案例 08 线程常用属性.currentThread，.enumerate:activeCount:thr.setName，thr.getName

import time
import threading

def loop1():
    print("start loop1 at:",time.ctime())
    time.sleep(6)
    print("end loop1 at:",time.ctime())
def loop2():
    print("start loop2 at:",time.ctime())
    time.sleep(1)
    print("end loop2 at:",time.ctime())
def loop3():
    print("start loop3 at:",time.ctime())
    time.sleep(5)
    print("end loop3 at:",time.ctime())
    
def main():
    print("starting......",time.ctime())
    
    t1 = threading.Thread(target=loop1,args=())
    t1.setName("TH_1")
    t1.start()
    
    t2= threading.Thread(target=loop2,args=())
    t2.setName("TH_2")
    t2.start()
    
    t3= threading.Thread(target=loop3,args=())
    t3.setName("TH_3")
    t3.start()
    
    time.sleep(5.5)
    for thr in threading.enumerate():
        print("正在运行的线程的名字是：{}".format(thr.getName()))
    print("正在巡行的子线程的数量是：{}".format(threading.activeCount()))
    print("All done")
    
if __name__ =="__main__":
    main()
    while True:
        time.sleep(1)#是几无所谓
```

```python
# 案例10 工业风
import threading
from time import ctime,sleep


class ThreadFunc:
    def __init__(self,name):
        self.name = name
    def loop(self,nloop,nsec):
        print("start loop",nloop,"at",ctime())
        sleep(nsec)
        print("Done loop",nloop,"at",ctime())
def main():
    print("Starting......",ctime())
    
    t = ThreadFunc("loop")
    t1 = threading.Thread(target=t.loop,args=("LOOP1",4))
    t2 = threading.Thread(target=ThreadFunc("loop").loop,args=("LOOP2",2))
    
    t1.start()
    t2.start()
    
    t1.join()
    t2.join()
    
    print("Done.......",ctime())
if __name__ == "__main__":
    main()
    
```

```python
# 案例12 上锁厚 lock.acquier() 解锁：release()
# 案例11，还未上锁时

import threading
import time

sum = 0
loopSum = 1000000
lock = threading.Lock()


def myAdd():
    global sum,loopSum
    for i in range(1,loopSum):
        lock.acquire()
        sum += 1
        lock.release()
def myMinu():
    global sum ,loopSum
    for i in range(1,loopSum):
        lock.acquire()
        sum -= 1
        lock.release()
        
def main():
    print("starting ....",sum)
    
    t1 = threading.Thread(target=myAdd,args=())
    t2 = threading.Thread(target=myMinu,args=())
    t1.start()
    t2.start()
    
    t1.join()
    t2.join()
    
    print("ending......",sum)
    
main()
```

```python
# 案例14 死锁问题 acquire release
import threading 
import time

lock1 = threading.Lock()
lock2 = threading.Lock()

def func1():
    print("func1 starting......")
    lock1.acquire()
    print("func1 申请了 锁lock1.......")
    time.sleep(2)
    print("func1 等待了 锁lock2......")
    lock2.acquire()
    print("func1 申请了 锁lock2....;..")
    lock2.release()
    print("func1 释放了 锁lock2......")
    lock1.release()
    print("func1 释放了 锁lock1......")
    
def func2():
    print("func21 starting......")
    lock2.acquire()
    print("func2 申请了 锁lock2.......")
    time.sleep(4)
    print("func2 等待了 锁lock1......")
    lock1.acquire()
    print("func2 申请了 锁lock1....;..")
    lock1.release()
    print("func2 释放了 锁lock1......")
    lock2.release()
    print("func2 释放了 锁lock2......")
    
t1 = threading.Thread(target=func1,args=())
t2 = threading.Thread(target=func2,args=())

t1.start()
t2.start()

t1.join()
t2.join()
```

```python
# 案例15 解决锁的等待问题
import threading 
import time

lock1 = threading.Lock()
lock2 = threading.Lock()

def func1():
    print("func1 starting......")
    lock1.acquire(timeout=2)
    print("func1 申请了 锁lock1.......")
    time.sleep(2)
    print("func1 等待了 锁lock2......")
    rst = lock2.acquire(timeout=2)
    if rst:
        print("func1 申请了 锁lock2....;..")
        lock2.release()
        print("func1 释放了 锁lock2......")
    else:
        print("func1 注定申请不到锁lock2")
    lock1.release()
    print("func1 释放了 锁lock1......")
    
def func2():
    print("func21 starting......")
    lock2.acquire()
    print("func2 申请了 锁lock2.......")
    time.sleep(4)
    print("func2 等待了 锁lock1......")
    lock1.acquire()
    print("func2 申请了 锁lock1....;..")
    lock1.release()
    print("func2 释放了 锁lock1......")
    lock2.release()
    print("func2 释放了 锁lock2......")
    
t1 = threading.Thread(target=func1,args=())
t2 = threading.Thread(target=func2,args=())

t1.start()
t2.start()

t1.join()
t2.join()
```

```python
# 案例16 允许一个资源最多由几个多线程同时使用，semphore
#  threading.currentThread().getName() + "get semaphore"

import threading
import time 

semaphore = threading.Semaphore(3)

def func():
    if semaphore.acquire():
        print(threading.currentThread().getName() + " get semphore")
        time.sleep(4)
        semaphore.release()
        print(threading.currentThread().getName() + " release semphore")
        
for i in range(8):
    t = threading.Thread(target=func)
    t.start()
```

```python
# 案例22 JoinableQueue task_done完成后给个反馈，daemon守护进程，get拿出来，put放进去
import multiprocessing
import time

def consumer(input_q):
    print("start consumer in :",time.ctime())
    while True:
        item = input_q.get()
        input_q.task_done()
        print("pull", item, " in here")
    print("end time is:",time.ctime())

def produce(sequence,output_q):
    print("start produce in :",time.ctime())
    for item in sequence:
        output_q.put(item)
        print("put",item,"into q")
    print("all produce")
    
if __name__ == "__main__":
    q = multiprocessing.JoinableQueue()
    
    cons_p = multiprocessing.Process(target=consumer,args=(q,))
    cons_p.daemon = True
    cons_p.start()
    
    sequence = [1,2,3,4]
    produce(sequence,q)
    q.join()
```

```python
# 案例23 哨兵问题
import multiprocessing
from time import ctime

# 设置哨兵问题
def consumer(input_q):
    print("Into consumer:", ctime())
    while True:
        # 处理项
        item = input_q.get()
        # if item == None:
        if not item:
            break
        print ("pull", item, "out of q") # 此处替换为有用的工作
        input_q.task_done() # 发出信号通知任务完成
    print ("Out of consumer:", ctime()) ##此句未执行，因为q.join()收集到四个task_done()信号后，主进程启动，未等到print此句完成，程序就结束了


def producer(sequence, output_q):
    # JoinableQueue：work干活，干完了给反馈
    print ("Into procuder:", ctime())
    for item in sequence:
        output_q.put(item)
        print ("put", item, "into q")
    print ("Out of procuder:", ctime())



# 建立进程
if __name__ == '__main__':
    q = multiprocessing.JoinableQueue()
    # 运行消费者进程
    cons_p = multiprocessing.Process (target = consumer, args = (q,))
    cons_p.daemon = True
    cons_p.start()

    # 生产多个项，sequence代表要发送给消费者的项序列
    # 在实践中，这可能是生成器的输出或通过一些其他方式生产出来
    sequence = [1,2,3,4]
    producer(sequence, q)
    q.put(None)
    # 等待所有项被处理
    cons_p.join()
```
